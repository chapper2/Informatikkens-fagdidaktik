Hvilke egenskaber og hvilken syntaks har en teksstreng ("string") typisk i programmering (Java/Javascript/Python).

Opgave 1:
Lav i Scratch en kode det tager et ord som input og lgger det i variablen svar?

Opgave 2:
F ved hjlp af indekset for en tekststreng udskrevet det frste og sidste bogstav af dit svar?

Prsentation/ Oplg 
Hvad er et palindrom og hvordan kan sammenligningsoperatorer og variable af typen boolean bruges?

Eleverne skal undersge det abstrakte koncept palindrom, og overveje/diskutere hvilke betingelser der skal vre opfyldt for at en teksstreng kan kaldes et palindrom (Refine de skal g fra det abstrakte til det konkrete)

Opgave 3: 
Forklar med pseudokode/tekst Hvordan man laver et tjek af om det frste og det sidste bogstav i en teksstreng er det samme, og hvad skal der ske, hvis ikke de er det samme (Extend)?

Lav derefter en funktion i scratch der kan tilfjes koden fra opgave 1 og 2 og som sammenligner frste og sidste bogstav i teksstrengen fra de to frste opgaver. Funktionen skal returnere Sandt eller falsk.

Opgave 4:
Lav det sledes at denne operation gentages ind mod midten? (restrukturering af koden fra opgave 1 og 2)
    Hvad hvis det er ulige antal bogstaver
Find ud af hvor mange tjek, der skal laves for at tjekke om det er et palindrom.

Opgave 5: 
Srg for at det bliver udskrevet om ordet er et palindrom eller ej.
Indst et break som stopper tidligt, s snart et tjek viser, at det ikke er palindrom. (optimering)
